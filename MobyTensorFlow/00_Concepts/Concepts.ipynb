# 1. Inteligencia Artificial

## 1.1. Historia

### 1.1.1. Génesis de la inteligencia artificial (1943-1955)

Los autores que han sido reconocidos como primeros en abordar un trabajo de IA han sido Warren McCulloch y Waltter Pitts en 1943. Para su trabajo partieron de tres fuentes: el funcionamiento de las neuronas, el análisis formal de lógica proposicional y la teoría de la computación. Estos propusieron un modelo constituido por neuronas artificiales. Poco después Donald Hebb, en 1949, propuso una regla para la modificación de las intensidades de las conexiones entre neuronas. 

Como se puede ver, en estos años empiezan a surgir trabajos que se pueden caracterizar como de IA. La primera persona que articuló una primera visión de la IA fue Alan Turing en su artículo “Computing Machinery and Intelligence” en 1950, donde introdujo conceptos como el aprendizaje automático o la prueba de Turing.

Un año más tarde, Marvin Minsky y Dean Edmonds, dos estudiantes graduados en el departamento de matemáticas de Princeton, construyen el primer computador a partir de una red neuronal en 1951.

### 1.1.2. Nacimiento de la inteligencia artificial (1956)

Princeton acogió a una de las figuras más importantes de la IA como fue John McCarthy. Después de graduarse, McCarthy se trasladó al Dartmouth College. Fue allí donde, junto a otros investigadores como Minsky o Claude Shannon, organizaron un taller en el verano de 1956, al cual acudieron asistentes tan prestigiosos como Arthur Samuel de IBM, Oliver Selfridge del MIT o Allen Newell y Herbert Simon del Carnegie Tech. 

Este taller no produjo avances demasiado importantes (aparte del consenso de adoptar el nombre de Inteligencia Artificial para este campo), pero puso en contacto a figuras importantes de este nuevo campo. En los años posteriores, fueron ellos los que lo dominaron.

### 1.1.3. Entusiasmo inicial, grandes esperanzas (1952-1969)

Durante los primeros años de la IA, se consiguieron grandes éxitos. Relativos, pues los computadores eran bastante primitivos y por aquel entonces se les consideraba únicamente herramientas para realizar cálculos aritméticos, pero éxitos al fin y al cabo. Poco a poco los computadores empezaban a realizar diversas tareas. Por ejemplo, el SRGP (Sistema de Resolución General de Problemas) de Newell y Simon, el DTG (Demostrador de Teoremas de Geometría) de Herbert Gelernter o los diversos programas para el juego de las damas que escribió Arthur Samuel.

Por su parte, John McCarthy realizó tres grandes contribuciones en 1958: el lenguaje de alto nivel Lisp, junto a otros investigadores inventaron el tiempo compartido y un programa que podría considerarse el primer sistema de IA completo llamado el Generador de Consejos.

Minsky supervisó el trabajo de una serie de estudiantes cuyo objetivo era resolver problemas limitados que requerían de inteligencia. Estos dominios concretos y limitados se conocen como micromundos. Un ejemplo de estos fue el programa STUDENT de Daniel Bobrow (1967) que era capaz de resolver problemas de álgebra, aunque el más famoso era el micromundo de los bloques.

### 1.1.4. Una dosis de realidad (1966-1973)

Debido a estos avances, los investigadores de IA auguraban grandes éxitos en los próximos años. Por ejemplo, Herbert Simon decía que en diez años un computador llegaría a ser campeón de ajedrez o que podría demostrar un importante teorema matemático. Aunque no estaba equivocado, estas predicciones no se cumplieron hasta 40 años después. En esta época surgieron diversos problemas que llevaron a que los gobiernos retiraran muchas de las subvenciones que se dedicaban a este campo. 

Un caso ejemplo de esto fue cuando se intentó traducir textos científicos rusos mediante traducción automática. En vísperas del lanzamiento del Sputnik (1957) se traducen frases rusas como “el espíritu es fuerte pero la carne es débil” por “el vodka es bueno pero la carne está podrida”. En 1966 un comité declaró que “no se ha logrado obtener ninguna traducción de textos científicos generales ni se prevé obtener ninguna en un futuro inmediato” lo que llevó al cancelamiento de todo tipo de patrocinio y ayudas del gobierno estadounidense a este tipo de proyectos.

En 1969, el libro de Minsky y Papert llamado “Perceptrons” demostró que aunque era posible conseguir que los perceptrones (una red neuronal simple) aprendiesen cualquier cosa que pudiesen representar, la capacidad de representación que poseían era muy limitada. Esto llevó a la retirada de la mayoría de fondos de investigación destinados a las redes neuronales.

Otro problema era las miles de horas de CPU que los sistemas necesitaban. Esta fue una de las principales críticas que se le hicieron al campo de la IA en el informe de Lighthill (1973). El gobierno británico se basó en este informe para retirar las ayudas a las investigaciones sobre IA.

### 1.1.5 Sistemas basados en el conocimiento: ¿clave del poder? (1969-1979)

Durante estos años surgieron los primeros sistemas expertos. El conocimiento de estos sistemas estaba basado en grandes cantidades de reglas de propósito particular.

Un caso particular de esto fue el programa MYCIN, hecho para el diagnóstico de infecciones sanguíneas. Este programa era capaz de hacer diagnósticos tan buenos como los de un experto y mucho mejores que médicos recién graduados. Las más de 450 reglas que poseía se obtuvieron a partir de extensas entrevistas a expertos, que las habían obtenido de libros de texto, otros expertos y experiencia directo en casos prácticos.

Poco a poco la necesidad de sistemas capaces de resolver problemas del mundo real fue en aumento. Esto, unido a la comprobada eficacia de sistemas expertos, hizo que su demanda fuera aumentando.

### 1.1.6. La IA se convierte en una industria (desde 1980 hasta el presente)

R1 fue el primer sistema experto comercial que tuvo éxito. Este programa se usaba en la elaboración de pedidos de nuevos sistemas informáticos. La compañia donde comenzó su actividad, Digital Equipment Corporation (McDermott, 1982), ahorraba alrededor de 40 millones de dólares al año.

Pocos años después, las compañías más importantes de Estados Unidos contaban ya con su propio grupo de IA enfocado en el uso o investigación de sistemas expertos.  Las ayudas y subvenciones al campo de la IA volvieron con fuerza, lo que provocó que la industria de la IA creciera rápidamente.

### 1.1.7. Regreso de las redes neuronales (desde 1986 hasta el presente)

El campo de las redes neuronales se había abandonado por la informática de manera general a finales de los 70, aunque el trabajo continuó en otros campos como la psicología o la física.

El mayor impulso se produjo a mediados de los 80, cuando cuatro grupos distintos reinventaron el algoritmo de aprendizaje de retroalimentación. Este algoritmo se aplicó en informática y la psicología a diversos problemas de aprendizaje. La publicación de los resultados (1986) tuvo una gran difusión y suscitaron gran entusiasmo.

### 1.1.8. IA se convierte en una ciencia (desde 1987 hasta el presente)

El contenido y metodología de trabajo en el campo de la inteligencia artificial ha sufrido una revolución en los últimos años. Es más común el desarrollo de teorías ya existentes que proponer nuevas teorías, usar como base evidencias experimentales y demostrar la utilidad de las aplicaciones en el mundo real.

Además, para que se acepten, las hipótesis hechas en el campo se deben someter a experimentos empíricos, y los resultados deben analizarse estadísticamente. Gracias al uso de internet, es posible compartir repositorios de datos de prueba y código con el fin de contrastar experimentos.

La  mejor comprensión de los problemas y su complejidad, sumado a un incremento en la sofisticación de las matemáticas ha facilitado el desarrollo de una agenda de investigación y de métodos más robustos.

Esta formalización y especialización ha llevado también a la que áreas como la visión y la robótica se vayan aislando de la «rama central» de la IA. El concepto de IA como diseño de agentes racionales facilita la unificación de los distintos subcampos.

### 1.1.9. Emergencia de los agentes inteligentes (desde 1995 hasta el presente)

Debido a que el progreso en los distintos subcampos de la IA  era llevado a buen ritmo, los investigadores retomaron el trabajo en el problema del “agente total”. Unos de los entornos más importantes para estos agentes es Internet. El uso de sistemas de IA se ha convertido en algo común en la web, tanto que el sufijo “-bot” se ha introducido en el lenguaje común. 

Las tecnologías de IA hoy en día son muy usadas en motores de búsqueda, sistemas de recomendaciones, etc.

### 1.1.10 La disponibilidad de enormes cantidades de datos (2001 - actualidad)

A lo largo de la historia de las ciencias de la computación se ha hecho énfasis en los algoritmos, pero recientes trabajos sugieren que para la solución de muchos problemas tiene más sentido preocuparse por los datos que por el tipo de algoritmo a usar.

Es decir, que un algoritmo mediocre alimentado con 100 millones de datos puede funcionar mejor que el mejor algoritmo alimentado con 1 millon de datos. Debido a que en los últimos años el almacenamiento y disponibilidad de datos ha mejorado enormemente, los sistemas de IA se han convertido en piezas fundamentales en todo tipo de industria.


## 1.2. Concepto y descripción
¿Que es la IA? El test de Turing

Definirla es una tarea que cuesta trabajo, aunque Alan Turing propuso la Prueba de Turing. Esta prueba se desarrollo con el fin de proporcionar una definición de inteligencia. La prueba esta basada en la incapacidad de diferenciar entre entidades inteligentes y seres humanos. Un sistema supera la prueba si la persona que lo evalua es incapaz de decir si las respuestas a una serie de preguntas dadas son proporcionadas por una persona o no.

## 1.3. Aproximaciones
A lo largo de la historia se ha definido desde distintas aproximaciones según como actuan o como piensan. Algunas de las definiciones que se han dado de la IA se podrían clasificar según realicen estas acciones, si las realizan de forma humana o de forma racional.

### 1.3.1 Pensamiento
#### 1.3.1.1 Pensando de forma humana 

"El nuevo y excitante esfuerzo de hacer que los computadores piensen… máquinas con mentes, en el más amplio sentido literal". (Haugeland, 1985)
"La automatización de actividades que vinculamos con procesos de pensamiento humano, actividades como la toma de decisiones, resolución de problemas, aprendizaje…" (Bellman, 1978)

Esta aproximación busca que los programas piensen de forma humana. Para lograr esto es necesario determinar como piensan los humanos y penetrar en el funcionamiento de la mente. El campo de la ciencia cognitiva (donde convergen modelos computacionales y técnicas de psicología) elabora teorías precisas y verificables sobre el funcionamiento de la mente humana. 

#### 1.3.1.2 Pensando de forma racional

"El estudio de las facultades mentales mediante el uso de modelos computacionales". (Charniak y McDermott, 1985)
"El estudio de los cálculos que hacen posible percibir, razonar y actuar". (Winston, 1992)

Aristóteles intentó codificar "la forma correcta (racional) de pensar". Buscaba llegar a conclusiones correctas partiendo de ciertas premisas, a través de esquemas de estructuras de argumentación. Esto fue el inicio del campo de la lógica.

Ya en 1965 existían programas que era capaces de resolver problemas resolubles descritos en notación lógica. En IA, este enfoque busca construir sistemas inteligentes a partir de este tipo de programas.

### 1.3.2 Actuacion
#### 1.3.2.1 Actuando de forma humana

"El arte de desarrollar máquinas con capacidad para realizar funciones que cuando son realizadas por personas requieren de inteligencia".
(Kurzweil, 1990)
"El estudio de cómo lograr que los computadores realicen tareas que, por el momento, los humanos hacen mejor". (Rich y Knight, 1991)

En este enfoque se busca construir sistemas que posean las siguientes capacidades:

* Representación del conocimiento: para almacenar lo que se conoce o siente
* Razonamiento automático: Usar el conocimiento que posee para responder preguntas y llegar a nuevas conclusiones
* Aprendizaje automático: para adaptarse a nuevas circunstancias y para detectar y extrapolar patrones.
* Procesamiento de lenguaje natural: que le permita comunicarse satisfactoriamente
* Visión computacional: para percibir objetos.
* Robótica: para manipular y mover objetos.

Estas capacidades "huamanas", que son las que se busca replicar, definen las seis disciplinas que abarcan la mayor parte de la IA.

#### 1.3.2.2 Actuando de forma racional

"La Inteligencia Computacional es el estudio del diseño de agentes inteligentes". (Poole et al., 1998)
"IA… está relacionada con conductas inteligentes en artefactos". (Nilsson, 1998)

Un agente, en el campo de la informática debe de tener atributos como: capacidad de razonar, controlarse de forma autónoma, percebir su entorno, adaptarse a los cambios y conseguir objetivos diferentes. 

Este enfoque busca obtener agentes racionales. Un agente racional es el que usando los atributos que posee, actua con la intención de alcanzar el mejor resultado posible o, al menos, el mejor resultado posible cuando hay una incertidumbre. Esta forma de actuar se diferencia de la humana en que esta es muy específica, ya que es producto de un proceso evolutivo complejo que está lejos de llevarnos a la perfección.

## 1.4. Casos de uso

Actualmente la IA es capaz de realizar muchas tareas en distintos subcampos:

Planificación autónoma: por ejemplo el programa de la NASA Agente REmoto. Este programa se encarga de la planificación de las operaciones de una nave espacial desde abordo.

Juegos: Por ejemplo AlphaStar, creado por Google. Este sistema derrotó a varios jugadores profesionales en el complicado videojuego de estrategia Starcraft.Aparte de este, el ejemplo más famoso es el del Deep Blue de IBM. Este sistema derrotó al campeón del mundo de ajedrez, Garry Kasparov con un resultado de 3.5 a 2.5. Kasparov dijo que habia percibido un "nuevo tipo de inteligencia al otro lado del tablero. 

Control autónomo: como ALVINN, un sistema de visión por computador que es capaz de dirigir un coche de forma que siguiese una línea. En uno de los viajes de prueba, de 2950 millas por otdo Estados Unidos, fue capaz de controlar la dirección del vehículo en el 98% del trayecto

Diagnosis: Muchos programas de diagnóstico basádos en el análisis probabilistico han llegado a niveles muy parecidos al de los médicos. No solo eso, sinó que algunos sistemas son capaces de indicar cómo han realizado y generado el diagnóstico, indicando los factores en los que basan sus decisiónes.

Planificación Logística: uno de los casos más conocidos es el de la herramienta DART (Cross y Walker, 1994) durante la crisis del Golfo Pérsico. Esta herramienta automatizaba la planificación y organización logística de más de 50.000 vehículos, su carga y personal, teniendo en cuenta los puntos de partida y destino, rutas, etc.

Robótica: actualmente muchos cirujanos usan asistentes robot en operaciones de microcirugía. Un ejemplo es HipNav

Procesamiento de lenguaje y resolución de problemas: como Prover B, un programa informático que resuelve crucigramas mejor que la mayoría de humanos.

La lista es enorme y podríamos continuar enumerando casos de uso, pero estos ejemplos bastan para ver la importacia que ha cobrado la IA en nuestros días.


# 2. Business Inteligence

## 2.1 Concepto

El termino de Business Inteligence hace referencia al uso de tecnologías y aplicaciones que permiten reunir, depurar y trabajar los datos en una empresa con el objetivo de convertirlos en conocimiento para dar soporte a la propia empresa y su toma de decisiones 

## 2.2 Data Warehouse

Data Warehouse es un almacén electrónico, donde las empresas guardan y mantienen los datos e información que reunen los diversos sistemas de una empresa. Estos datos deben almacenarse de forma segura, fiable y fácil de administrar. Normalmente se aloja en un servidor corporativo o cada vez más, en la nube. A estos datos posteriormente se le suelen aplicar técnicas de Business Inteligence para tomar decisiones estratégicas en los negocios.

## 2.3 Roles
### 2.3.1 BI Analyst

El BI Analyst es el encargado de analizar y crear modelos útiles para el uso de los datos que posee la empresa. Monitorizan y trabajan con datos y métricas para poder dar soporte a la toma de decisiones de la empresa . Suelen trabajar junto al BI Engineer.

### 2.3.2  BI Engineer

El BI engineer es el encargado de asegurar que los datos son precisos, congruentes y de fácil acceso. Ellos proveen la infraestructura sobre la que sostener los datos de la empresa. Suele trabajar junto al BI Analyst.

# 3. Data Science
## 3.1 Concepto
Es la ciencia centrada en el estudio de datos con el objetivo de descubrir patrones a partir de estos usando distintas herramientas. Tambien se encarga de todo lo relacionado con la limpieza, preprocesamiento y análisis de los datos.

## 3.2 Data Analysis

Actividades de las personas que tienen como objetivo obtener una visión más clara de los datos e información que se posee. 

## 3.3 Data Analytics

Es la ciencia encargada de examinar datos con el objetivo de descubrir nuevos conocimientos y conclusiones a partir de estos.

## 3.4 Data Mining
La minería de datos es una parte muy grande del mundo de data science, ya que es la encargada de sacar información a partir de los datos para poder obtener resultados de valor,  como predicciones, clasificaciones, etc.

## 3.5 Roles
### 3.5.1 Data Analyst

Para este rol, se necesitan conocimientos de estadistica, además de capacidad para manejar y mostrar de forma efectiva la información.

### 3.5.2 Data Engineer

Este rol implica la preparación de los datos, para lo que necesitan tener conocimientos técnicos con distintas APIs. Se encargan del desarrollo, construcción, testeo y mantenimiento de la arquitectura que da soporte a los datos.

### 3.5.3 Data Scientist

Son los encargados de analizar e interpretar grandes cantidades de datos. Tienen conocimiento de estadística, machine learning, acondicionamiento y preparación de datos, etc.

# 4. Big Data

Las grandes cantidades de información que se manejan hoy en día no se pueden tratar de forma efectiva con las aplicaciones tradicionales. El big data como tal, hace referencia a las técnicas para usar grandes volúmenes de datos (estructurados o no), su almacenamiento y procesamiento para poder extraer valor.
