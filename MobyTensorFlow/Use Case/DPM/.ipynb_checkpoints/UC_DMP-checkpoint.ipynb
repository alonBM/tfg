Importamos las librerías y funciones necesarias

```python
import tensorflow as tf
import pandas as pd
import skmob

from skmob.measures.individual import home_location
from skmob.io.file import load_geolife_trajectories
```


Cargamos el dataset usando la función 'load_geolife_trajectories' que nos proporciona skmob. Esta función se encarga de cargar los datos de GeoLife Trajectories, solo hay que indicarle el path del directorio Data, y los usuarios que queremos cargar. Es opcional pasarle algunos parámetros de filtrado y compresión.

```python
path = "/home/alonso/Documentos/tfg/Geolife_Trajectories/"
traj_df = load_geolife_trajectories(path, user_ids=['000'],compress_kwargs={'spatial_radius_km': 0.5})
```

```python
print(type(traj_df))
traj_df.head()
```

```python

df = pd.DataFrame(traj_df)
print(type(df))
print(df.count())
df.head()
```

```python

home_loc_df = home_location(df)
home_loc_df.head()
```

```python
mergedStuff = pd.merge(home_loc_df, df, on=['lng','lat'])
mergedStuff.head()
```

```python
df['at_home'] = 0


df.loc[df['lat'].isin(home_loc_df['lat']), 'at_home'] = 1


df[df['at_home'] == 0].count()
```
