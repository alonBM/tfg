# Pasos de instalación


## 1. Instalación de Anaconda

Instalamos Anaconda desde la web oficial https://www.anaconda.com/products/individual

## 2. Clonamos el repositorio de scikit-mobility

Vamos al repositorio de Github de scikit-mobility y hacemos click en "Clone or download". Lo descargamos y descomprimimos donde queramos.

## 3. Abrimos la shell de anaconda

Abrimos la terminal llamada anaconda powershell prompt

## 4. Creamos un nuevo entorno y lo activamos

Creamos el entorno en el que trabajaremos e instalaremos las librerias necesarias con el siguiente comando:

    conda create --name myenv

Para activar este entorno usamos:

    conda activate myenv 

## 5. Instalamos tensorflow usando conda

Instalaremos la versión 2.1.0 de Tensorflow, que instalará la versión 3.7.7 de Python con el comando:

    conda install -c anaconda tensorflow

## 6. Opcional (Solo en windows)

Es necesaria la instalación de la libreria fiona con el siguiente comando:

    conda install -c conda-forge fiona=1.8.6

## 7. Corremos el script de instalación de scikit-mobility

Está en la carpeta del repositorio que hemos clonado. El archivo es "setup.py". Nos movemos en la terminal hasta la carpeta y ejecutamos:

    python setup.py install

## 8. Instalamos jupyterlab 

Instalamos el entorno de desarollo jupyterlab:

    conda install -c conda-forge jupyterlab

## 9. Desactivamos el entorno

Salimos del entorno usando:

    conda deactivate

## 10. Forma de trabajo

Ya tendriamos el entorno con todo lo necesario. Para trabajar usando las librerias, activamos el entorno y abrimos jupyterlab con los siguientes comandos

    conda activate my env

    jupyter lab

Cuando hayamos acabado, cerramos jupyterlab y desactivamos el entorno con el comando:

    conda deactivate
