# 1. Problema a abordar 

El problema se centrará en usar y tratar datos relacionados con la movilidad para poder hacer predicciones. Concretamente, intentaremos predecir cuando una persona está en casa o no.

## Datos
Los datos que usaremos han sido recolectados en el proyecto Geolife (Microsoft Research Asia). Corresponden a datos de 182 usuarios durante un periodo de 5 años (desde el 2007 hasta agosto de 2012). Los datos son trayectorias, representadas por varios puntos, cada uno con información de la latitud, la longitud, la altitud y el momento en el que se tomó esta información. Contiene un total de 17621 trayectorias, que recorren una distancia total de 1.292.951 kilometros y una duración total de 50.176 horas. Los datos se ha recolectado usando distintos sitemas de GPS. Los datos de este dataset, estan distribuidos en más de 30 ciudades de china, algunas ciudades de Estados Unidos y Europa. Aqui un ejemplo de las trayectorias:

Formato de los archivos PLT:
* Columna 1: Latitud en grados decimales.
* Columna 2: Longitud en grados decimales.
* Columna 3: Puestos a 0 para este set de datos.
* Columna 4: Altitud en pies.
* Columna 5: Fecha como decimales que han pasado desde el 12/30/1899.
* Columna 6: Fecha como string.
* Columna 7: Hora como string.

Además, las 6 primeras lineas del archivo contienen información que puede ser ignorada.

Ejemplo: 39.906631,116.385564,0,492,40097.5864583333,2009-10-11,14:04:30

## Lenguaje

## Librerías 
Tensorflow: es una librería de código, que proporciona un alto rendimiento de computación numérica. Trabaja con tensores, que se trata de una generalización de los vectores. Su arquitectura es flexibe y permite computación tanto en CPUs, como en GPUs. En este caso, la usaremos para crear un modelo de red neuronal, ya que también proporciona soporte para Machine Learning y Deep Learning.
