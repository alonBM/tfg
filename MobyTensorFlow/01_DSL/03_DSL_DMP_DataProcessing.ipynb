# 3. Procesado de datos

En esta fase se preparan los datos para su posterior uso. Lo primero sería entender la información que poseemos, decidir que hacer en caso de que haya valores no válidos o vacíos, etc. También es necesaria la limpieza de la información, su preparación o su posible reescalado.

Es aconsejable escribir funciones en vez de hacerlo manualmente por las siguientes razones:

* Permite reproducir las transformacions facilmente en cualquier dataset
* Permite construir gradualmente una libreria de transformaciones que puedes rehusar en futuros proyectos.
* Permite facilmente su posterior uso en tu propio sistema para tratar datos antes de introducirlos a tu sistema
