6. Resultados

En esta parte debemos presentar la solución. Hay que resaltar lo aprendido, que ha funcionado y que no, que suposiciones hemos realizado, cuales son las limitaciones del sistema y documentarlo todo. 

En caso de que se decidiera poner en producción, habría que realizar test además de monitorizar  y evaluar el rendimiento en vivo del sistema, si es que este cae. Evaluar el rendimiento de sistema requerirá de análisis humano por lo que es conveniente que este análisis lo hagan expertos en el campo

Es necesario tambien que se evalue la calidad de los datos que entran al sistema. A veces el rendimiento del sistema baja simplemente por la mala calidad de los datos que recibe (como sensores que envian valores aleatorios, otro sistema que alimenta el nuestro se ha degradado, etc)
