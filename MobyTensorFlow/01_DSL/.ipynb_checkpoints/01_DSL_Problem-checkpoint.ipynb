# 1. Definir y entender el problema

## Enmarcar el problema

Las primeras preguntas son obvias, pero fundamentales.

¿Cual es el objetivo? Debemos tener esto muy en cuenta, ya que de acuerdo a él, eligiremos un modelo u otro, la medida de rendimiento elegiremos, etc.

¿Cuál es la solucion actual? Esta es la siguiente pregunta que deberiamos responder. Esto nos dará una referencia del método actual, así como posibles ideas para resolver el problema. 

Otras preguntas comunes podrían ser las siguientes:
¿Es supervisado, no supervisado o por refuerzo?
¿Es una tarea de clasificación, regresión u otra? 

## Seleccionar una medida de rendimiento

Hay distintas y variadas. Por ejemplo, una típica para problemas de regresiones es la raíz del error cuadrático medio en inglés. Para problemas de clasificación se suelen usar matrices de confusión por ejemplo.

## Verificar las suposiciones

Siempre es una buena práctica listar y verificar las decisiones y suposiciones que hemos tomados, de esta forma identificaremos y resolveremos problemas en una etapa temprana. La equivocación es posible, podemos enfocar el problema de forma erronea, y si no se identifica este error pronto, se podrían perder semanas de trabajo.
