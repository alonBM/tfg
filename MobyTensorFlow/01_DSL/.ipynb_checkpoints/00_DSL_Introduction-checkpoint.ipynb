# 1 Introduction


Dentro del Data Science Lifecyle Se pueden encontrar en el 99% de los casos los siguientes pasos:

1. Definir y entender el problema
2. Obtener los datos
3. Procesar los datos
4. Trabajar estadísticamente con los datos
5. Selección del modelo para los datos
6. Resultados

Estos procesos no son fijos, dependiendo de distintos autores o profesionales del sector pueden cambiar de nombre u orden. Además, es bastante común iterar entre estos procesos. A continuación, explicaremos en que consiste cada uno de los pasos que hemos numerado.

Finalmente, se ha estudiado y analizado un proyecto de ejemplo, para obtener un mayor entendimiento de cómo se trabaja en el campo. Este proyecto y su código se encuentra en el libro "Hands-On Machine Learning with Scikit-Learn and TensorFlow" de Aurélien Géron, donde explica de manera práctica los puntos anteriormente mencionados.
