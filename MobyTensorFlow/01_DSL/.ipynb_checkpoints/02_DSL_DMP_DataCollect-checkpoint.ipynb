
# 2. Obtener los datos

Normalmente, los datos estarán disponible en una base de datos, esparcida entre distintas tablas, documentos, etc. y necesitaremos autorización para acceder a ellas y familiarizarnos con el esquema de los datos.

Esta parte está más relacionada con los Data engineer de los que hablabmos antes y se podría profundizar bastante en ella. Juega un papel muy importante, ya que son los datos los que nos permiten poder realizar las predicciones o clasificaciones.
